```python
# ==========================================
# IMPORT LIBRARY
# ==========================================

import pandas as pd

# ==========================================
# DATA JUMLAH KASUS
# ==========================================

jumlah_kasus = pd.DataFrame({

    'Kabupaten/Kota': [

        'Agam',
        'Bukittinggi',
        'Dharmasraya',
        'Kepulauan Mentawai',
        'Lima Puluh Kota',
        'Padang',
        'Padang Panjang',
        'Padang Pariaman',
        'Pariaman',
        'Pasaman',
        'Pasaman Barat',
        'Payakumbuh',
        'Pesisir Selatan',
        'Sawahlunto',
        'Sijunjung',
        'Solok',
        'Solok Selatan',
        'Tanah Datar'

    ],

    'Jumlah_Kasus': [

        6,3,1,1,2,5,9,3,5,
        5,4,4,4,4,1,2,3,2

    ]

})

# ==========================================
# CLUSTER AWAL
# ==========================================

# Cluster:
# Rendah = 2
# Sedang = 1
# Tinggi = 0

jumlah_kasus['Cluster'] = [

    1,1,2,2,2,1,0,1,1,
    1,1,1,1,1,2,2,1,2

]

print("DATA AWAL")
print(jumlah_kasus)

# ==========================================
# ITERASI ULANG K-MEANS
# ==========================================

print("\n================================")
print("ITERASI ULANG")
print("================================")

# ==========================================
# MENGHITUNG CENTROID BARU
# ==========================================

centroid_baru = {}

for cluster in sorted(jumlah_kasus['Cluster'].unique()):

    anggota = jumlah_kasus[
        jumlah_kasus['Cluster'] == cluster
    ]['Jumlah_Kasus']

    centroid = anggota.mean()

    centroid_baru[cluster] = centroid

print("\nCENTROID BARU")

for c, nilai in centroid_baru.items():

    print(f"Cluster {c} = {nilai:.2f}")

# ==========================================
# MENENTUKAN CLUSTER BARU
# ==========================================

cluster_hasil = []

print("\n================================")
print("PERHITUNGAN JARAK")
print("================================")

for i in range(len(jumlah_kasus)):

    data = jumlah_kasus['Jumlah_Kasus'][i]

    # Hitung jarak
    d0 = abs(data - centroid_baru[0])
    d1 = abs(data - centroid_baru[1])
    d2 = abs(data - centroid_baru[2])

    # Simpan jarak
    jarak = [d0, d1, d2]

    # Cari cluster terdekat
    cluster_terdekat = jarak.index(min(jarak))

    cluster_hasil.append(cluster_terdekat)

    print(
        f"{jumlah_kasus['Kabupaten/Kota'][i]}"
        f" | Kasus = {data}"
        f" | C0 = {d0:.2f}"
        f" | C1 = {d1:.2f}"
        f" | C2 = {d2:.2f}"
        f" | Cluster Baru = {cluster_terdekat}"
    )

# ==========================================
# MENYIMPAN HASIL ITERASI
# ==========================================

jumlah_kasus['Cluster_Baru'] = cluster_hasil

# ==========================================
# CEK APAKAH CLUSTER BERUBAH
# ==========================================

print("\n================================")
print("HASIL ITERASI")
print("================================")

print(
    jumlah_kasus[
        [
            'Kabupaten/Kota',
            'Jumlah_Kasus',
            'Cluster',
            'Cluster_Baru'
        ]
    ]
)

# ==========================================
# CEK KONVERGEN / TIDAK
# ==========================================

if all(
    jumlah_kasus['Cluster']
    ==
    jumlah_kasus['Cluster_Baru']
):

    print("\nKESIMPULAN:")
    print("Cluster tidak berubah")
    print("K-Means sudah konvergen")

else:

    print("\nKESIMPULAN:")
    print("Masih ada perubahan cluster")
    print("Perlu iterasi berikutnya")
```
